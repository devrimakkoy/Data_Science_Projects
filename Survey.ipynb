```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go #learning the plotly library
import plotly.express as px
```

```python
#remove the top row(questions)
df_fin = df.iloc[1:,:]
df_fin.head()
```

```python
#get percent of null values in question
df_fin.isnull().sum() / df.shape[0]
```

```python
#create a dictionary for questions
Questions = {}
qnums = list(dict.fromkeys([i.split('_')[0] for i in df_fin.columns]))
qnums
```

```python
#add data for each question to key value pairs in dictionary
for i in qnums:
    if i in ['Q1','Q2','Q3']:
        Questions[i] = df_fin[i]
    else:
        Questions[i] = df_fin[[q for q in df_fin.columns if q.startswith(i)]]
```

```python
#Q1 histogram using px
fig = px.histogram(df_fin, x = 'Q1')
fig.show()
```

```python
#heatmap using px for Q1 & Q6 (age vs how many years they've been writing code)
fig = px.density_heatmap(df_fin, x='Q1', y='Q6', category_orders={'Q1':['18-21','22-24','25-29','30-34','35-39','40-44','45-49','50-54','55-59','60-69','70+'],
'Q6':['I have never written code','< 1 years','1-2 years','3-5 years','5-10 years','10-20 years','20+ years']})
fig.show()
```
